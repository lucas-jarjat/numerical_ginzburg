```python
from scipy.optimize import curve_fit
from numpy import *
import matplotlib.pyplot as plt
import inspect
%matplotlib inline
```

```python
file = "fit_sinus.dat"
data = loadtxt(file)
datax = data[:, 0]
datay = data[:, 1]
plt.plot(datax, datay, 'o')
```

```python
def frange(x, a=0.080, b=1/200, c=0.22, f0=0):
    return a*cos(b*2*pi*(x-f0))+c

X = linspace(-220, 220, 1000)
Y = frange(X, 0.080, 1/200, 0.22, 0)
plt.plot(datax, datay, 'o')
plt.plot(X,Y)
```

```python
param_opti = curve_fit(frange, datax, datay, p0 = [0.08, 1/200, 0.22, 0])[0]
Y = frange(X, *param_opti)
plt.plot(X,Y, color='g')
plt.plot(datax, datay, 'o', color='r')
```

```python
incer = sqrt(curve_fit(frange, datax, datay, p0 = [0.08, 1/200, 0.22, 0])[1][3][3])
print("f0 = {f0} +/- {incer} Hz".format(f0=param_opti[3], incer=incer))
```

```python
def my_curve_fit(f, xdata, ydata, p0={}):
    args, _, _, defaults = inspect.getargspec(f)
    defaults = list(defaults)
    for cle, val in p0.items():
        i = args.index(cle)
        defaults[i-1] = val
    popt, _ = curve_fit(f, xdata, ydata, p0=defaults)
    out = dict(zip(args[1:], popt))
    return out
```

```python
popt = my_curve_fit(frange, datax, datay, p0={})
print(popt)
```

```python
class FitSinus(object):
    _popt = None
    def __init__(self, temps, datay, frequence):
        self.temps = temps
        self.datay = datay
        self.frequence = frequence
        
        
    @staticmethod
    def fit_sinus(x, amplitude=1, offset=0, phase=0, frequence=10):
        return offset+amplitude*sin(2*pi*frequence*x+phase)   
    
    def fit(self):
        amplitude = (self.datay.max() - self.datay.min())
        offset = (self.datay.max() + self.datay.min())
        p0 = {'amplitude':amplitude, 'frequence':self.frequence, 'offset':offset}
        popt = my_curve_fit(self.fit_sinus, self.temps, self.datay, p0=p0)
        if popt["amplitude"]<0:
            popt["amplitude"] = -popt[amplitude]
            popt["phase"] += pi
        return popt
    
    def plot(self):
        t = linspace(self.temps.min(), self.temps.max(), 100)
        plt.plot(self.temps, self.datay)
        plt.plot(t, self.fit_sinus(t, **self.popt))
        
    
    @property
    def popt(self):
        if self._popt is None:
            self._popt = self.fit()
        return self._popt
    
    @property
    def phase(self):
        return self.popt["phase"]
    
    @property
    def amplitude(self):
        return self.popt["amplitude"]
```
